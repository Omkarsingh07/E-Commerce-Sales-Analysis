```python
import pandas as pd 
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
import plotly.colors as colors
pio.templates.default = "plotly_white"

data = pd.read_csv("sample.csv" , encoding = 'latin-1')
```

```python
#converting date coloumn 
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'],)
```

```python
data.info()
```

```python
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
```

```python
data.info()
```

```python
data.head()
```

```python
data['Order Month'] = data['Order Date'].dt.month
data['Order Year'] = data['Order Date'].dt.year
data['Order Day of Week'] = data['Order Date'].dt.dayofweek
```

```python
data.head()
```

# monthly sales analysis

```python
sales_by_month = data.groupby('Order Month')['Sales'].sum().reset_index()
```

```python
sales_by_month
```

```python
import plotly.express as px

fig = px.line(sales_by_month,
              x='Order Month',
              y='Sales',
              title='Monthly Sales Analysis')

fig.write_html("sales_analysis.html", auto_open=True)
```

# Sales by Category

```python
 sales_by_category = data.groupby('Category')['Sales'].sum().reset_index()
```

```python
sales_by_category
```

```python
import plotly.express as px

fig = px.pie(sales_by_category,
             values='Sales',
             names='Category',
             title='Sales by Category')

fig.update_traces(textposition='inside', textinfo='percent+label')
fig.write_html("sales_pie.html", auto_open=True)
```

# Sales Analysis by Sub Category

```python
sales_by_sub_category = data.groupby('Sub-Category')['Sales'].sum().reset_index()
```

```python
sales_by_sub_category
```

```python
import plotly.express as px
fig = px.bar(sales_by_sub_category,
              x='Sub-Category',
              y='Sales',
              title='Sales by Sub-Category')

fig.update_layout(xaxis_title="Category",
                  yaxis_title="Sales",
                  uniformtext_minsize=8,
                  uniformtext_mode='hide')

fig.write_html("sales_bar.html", auto_open=True)
```

# Profit by customer segment

```python
data.head()
```

```python
profit_by_segment = data.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
profit_by_segment
```

```python
import plotly.express as px
```

```python
fig =px.bar(profit_by_segment,
            x='Segment',
            y=['Sales','Profit'],
            title='Sales and Profit by Customer Segment',
            barmode='group')
fig.write_html("profit_by_segment.html", auto_open=True)
```

# Sales to Profit Ratio

```python
sales_by_segment = data.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
sales_by_segment['Sales_to_Profit_Ratio'] = sales_by_segment['Sales']/sales_by_segment['Profit']
print(sales_by_segment[['Segment','Sales_to_Profit_Ratio']])
```
